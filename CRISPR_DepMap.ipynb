```python
import umap
import numpy as np
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
%matplotlib inline

reducer = umap.UMAP()
```

```python
df_data = df.drop(['ModelID'], axis = 1)
df_data = StandardScaler().fit_transform(df_data)
```

```python
df_data.shape
```

```python
embedding = reducer.fit_transform(df_data)
embedding.shape
```

```python
plt.scatter(embedding[:, 0], embedding[:, 1], color = 'green', marker = '*')
plt.gca().set_aspect('equal', 'datalim')
plt.title("UMAP projections of the entire DepMAP Dataset", fontsize = 18)
plt.show()
```
